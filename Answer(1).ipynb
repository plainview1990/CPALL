```python
excel = ['01','02','03']
dfs = []
prefix = 'https://github.com/plainview1990/CPALL/raw/main/Quiz/'
dfs = [pd.read_excel(f'{prefix}{i}.xlsx')for i in excel]
```

```python
x1 = dfs[0][1:][['Zone','DATA 1','DATA 2','DATA 6']]
x2 = dfs[1][1:][['Zone','DATA 1','DATA 2','DATA 6']]
x3 = dfs[2][1:][['Zone','DATA 1','DATA 2','DATA 6']]
x1 #test dataframe x1 - x3
```

```python
x1 = x1.round(2)
x2 = x2.round(2)
x3 = x3.round(2)
x1 #test dataframe x1 - x3
```

```python
x1.loc[x1['DATA 2'] < 7,'DATA 2'] = 0
x2.loc[x2['DATA 2'] < 7,'DATA 2'] = 0
x3.loc[x3['DATA 2'] < 7,'DATA 2'] = 0
x1 #test dataframe x1 - x3
```

```python
x1.loc[x1.duplicated(['DATA 6'], keep=False), ['DATA 6']] = ''
x2.loc[x2.duplicated(['DATA 6'], keep=False), ['DATA 6']] = ''
x3.loc[x3.duplicated(['DATA 6'], keep=False), ['DATA 6']] = ''
x1 #test dataframe x1 - x3
```

```python
sum_data1 = {'Zone':'','DATA 1':x1['DATA 1'].sum(),'DATA 2':'','DATA 6':''}
sum_data2 = {'Zone':'','DATA 1':x2['DATA 1'].sum(),'DATA 2':'','DATA 6':''}
sum_data3 = {'Zone':'','DATA 1':x3['DATA 1'].sum(),'DATA 2':'','DATA 6':''}

x1 = x1.append(pd.DataFrame([sum_data1]))
x2 = x2.append(pd.DataFrame([sum_data2]))
x3 = x3.append(pd.DataFrame([sum_data3]))

# WARNING!!! Run only once. Otherwise, you have to clear all output and re-run again. 
```
